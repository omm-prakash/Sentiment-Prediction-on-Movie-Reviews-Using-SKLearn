Out of **13111** directors

4. Only **0.19%** of directors(i.e, 25) have movies more than **eighteen**.

From the above obervations directors can be encoded as `<=2`, `2< & <=6`, `6< & <=12`, `12< & <=18`, `>18`, `Joseph Brooks` & `Augusta Daniel` chategory. As we have only *16916* movies with *13111* directors, we can not label each directors indivisually. The encoding would be base on their *experience and popularity*.  


````python
def encode_director(n):
    if n<=2:
        return '<=2'
    elif n<=6:
        return '2< & <=6'
    elif n<=12:
        return '6< & <=12'
    elif n<=18:
        return '12< & <=18'
    elif n==100:
        return 'Augusta Daniel'
    elif n==208:
        return 'Joseph Brooks'
    else:
        return '>18'
    
director_data['category'] = director_data.movieid.apply(encode_director)
director_data.category.value_counts().to_frame()
````

````python
print('Encoding directors')
mdata['director_category'] = mdata.director.apply(lambda x: director_data.loc[x, 'category'])
mdata.director_category.value_counts().to_frame()
````

**directory_category_average_audienceScore**

````python
directory_category_average_audienceScore = director_data.groupby('category').audienceScore.mean().to_frame()
directory_category_average_audienceScore
````

````python
directors_categories = directory_category_average_audienceScore.index.to_list()
def get_average_audienceScore(movie):
    ndirectors = final_movie.loc[movie, directors_categories].sum()
    weighted_num = 0
    for category in directors_categories:
        director_score = directory_category_average_audienceScore.loc[category, 'audienceScore']
        ndirectors_category = final_movie.loc[movie,category]
        weighted_num += director_score*ndirectors_category
    return weighted_num/ndirectors
````

````python
def plot_venn2_diagram(l1, l2, 
                       set_colors = ('g','r'), 
                       set_labels = ('A', 'B'),
                       alpha = 0.8,
                       title = None,
                       show = True
                      ):
    n, m = len(l1), len(l2)
    inter = len(np.intersect1d(l1,l2))
    venn2((n-inter,m-inter,inter), 
          set_colors=set_colors,
          set_labels=set_labels,
          alpha=alpha,         
         )
    if title is not None:
        plt.title(title, color='b')
    if show:
        plt.show()
````

To confirm the dependence between these categorical features, Let's have **$\chi^2$ Test**. 

$$\chi^2 = \sum \frac {(O - E)^2}{E}$$
O: Obserbed frequency<br>
E: Expected frequency<br>

Ho: Features `rating(T/F)` & `sentiment` are independent. <br>
He: Features `rating(T/F)` & `sentiment` are dependent. 
$$\alpha = {0.05}$$

````python
sentiment_rating_count = train_movie_sentiment.groupby('sentiment').rating.value_counts().to_frame()
sentiment_rating_count['sentiment_freq'] = sentiment_rating_count.index.map(lambda x: train_movie_sentiment.sentiment.value_counts()[x[0]])
sentiment_rating_count['rating_freq'] = sentiment_rating_count.index.map(lambda x: train_movie_sentiment.rating.value_counts()[x[1]])
sentiment_rating_count['expected_freq'] = (sentiment_rating_count.sentiment_freq*sentiment_rating_count.rating_freq)/train_movie_sentiment.shape[0]
sentiment_rating_count
````

````python
# saving sentiment_rating_count for future use
with open('sentiment_rating_count.pkl', 'wb') as file:
    pkl.dump(sentiment_rating_count, file)
````

````python
alpha = 0.05
print('Result from Chi^2 test')
scipy.stats.chisquare(f_obs=sentiment_rating_count.rating, f_exp=sentiment_rating_count.expected_freq)
````

>From the above result it is clear that with `pvalue=6.971e-121`, which is less the $\alpha=0.05$. The null hypothesis would be **rejected**, i.e., **`rating(T/F)` & `sentiment` are dependent**. 

<br>
If we have rating then more likely that would be Negative and if we dont then more likly that would be Positive.

#### For Classification Problems

1. **f_classif**: This is the default scoring function for classification problems. It uses ANOVA F-value to evaluate the difference in means between classes. It's suitable when your target variable is categorical.
2. **chi2**: Chi-squared statistic measures the dependence between variables, which is useful for categorical features and categorical targets.
3. **mutual_info_classif**: This function measures the dependency between two random variables in the context of classification. It works well for both categorical and continuous features.

Hence taking **f_classif** as `score_func` in the `SelectKBest`, as we have counts and neumerical continuous  values as features.
